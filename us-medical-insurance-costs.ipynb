# U.S. Medical Insurance Costs

```python
#Importing csv file
import csv
```

```python
patient_age = []
patient_sex = []
patient_bmi = []
num_children = []
smoker_nonsmoker = []
patient_region = []
insurance_charges = []
```

```python
#Function for loading lists
def data_in_list (list_name, insurance_csv, column_name):
    with open ('insurance.csv') as insurance_data: #saves csv file 
        insurance_dict = csv.DictReader(insurance_data) #parses csv file, creates a dictionary
        for row in insurance_dict:                      #iterates through file rows in dictionary
            list_name.append(row[column_name])          #adds row column name to respective list created
        return list_name                                #returns list_name
     
```

```python
#calling fucntion to reveal data in lists
data_in_list(patient_age, 'insurance.csv', 'age')
data_in_list(patient_sex, 'insurance.csv', 'sex')
data_in_list(patient_bmi, 'insurance.csv', 'bmi')
data_in_list(num_children, 'insurance.csv', 'children')
data_in_list(smoker_nonsmoker, 'insurance.csv', 'smoker')
data_in_list(patient_region, 'insurance.csv', 'region')
data_in_list(insurance_charges, 'insurance.csv', 'charges')
```

```python
#1. Determining the average age of patients
total_age = 0
for age in patient_age:
    total_age += int(age)
    average_age = round(total_age/len(patient_age), 2)
print ("Average age: " + str(average_age))
```

```python
#2. Determining unique regions represented in the dataset
unique_regions = []
for region in patient_region:
    if region not in unique_regions:
        unique_regions.append(region)
print(unique_regions)
```

```python
#3. Determining where the majority of individuals are from
count_southwest = patient_region.count('southwest')
print ("southwest: " + str(count_southwest))
count_southeast = patient_region.count('southeast')
print ("southeast: " + str(count_southeast))
count_northwest = patient_region.count('northwest')
print ("northwest: " + str(count_northwest))
count_northeast = patient_region.count('northeast')
print ("northeast: " + str(count_northeast))
print("Majority of the individuals (728) come from the southeast region.")

```

```python
#4. Determing the cost difference between smokers and non_smokers
keys = insurance_charges
values = smoker_nonsmoker
smoking_diff = dict(zip(keys, values))
print(smoking_diff)
for key, value in smoking_diff.items():
    smoking_total = 0
    nonsmoking_total = 0
    if value == 'yes':
        smoking_total =+ float(key)
    else:
        nonsmoking_total += float(key)
insurance_diff = (smoking_total)- (nonsmoking_total)

print("The insurance_cost difference between smokers and non-smokers is: " + str(insurance_diff))
```

```python
#5. Calculating the average age of parents with at least 1 child.
age_children = list(zip(patient_age, num_children))
parent_age = []
total_age = 0
for age, child in age_children:
    if int(child) > 0:
        parent_age.append(age)
        total_age += int(age)
average_parent_age = round(total_age/len(parent_age),2)
print("The average age of parents with at least one child is: " + str(average_parent_age))
```
